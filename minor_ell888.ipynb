## Question 1: Missing feature problem - A quick literature review

In order to understand the missing value problem and available studies for possible solutions conducted a quick literature review. Findings of the review are suppariezed as follows.

### **Missing feature mechanism**

**MCAR**

Missing completely at random (MCAR). Rubin(1976) introduced the term "missing completely ar random" (MCAR) to describe the data where the complete cases are a random sample of the originally identified set of cases.

Since the complete cases are representative of the originally identified sample, inference based on only the complete cases are applicable to the larger sample and the target population.

Complete-case analysis for MCAR data provides results that are generalizable to the target population with one caveat - the estimates will be less precise than initially planned since a smaller number of cases are used for estimation.

**MAR**

Missing at random (MAR). Missing values are missing for reasons related to another variable, that is completely observed. For such cases Rubin(1976) uses the term "missing at random" (MAR) to describe data that are missing for reasons related to completely observed variables in the data set.

When data are MCAR or MAR, the response mechanism is termed ignorable, Ignorable response mechannisms are important because when they occur, resons for missing data in the analysis of the data can be ignored, and thus model-based methods can be used for missing data analysis.

**NMAR**

Not missing at random (NMAR). The pattern of data missingness is non-random and depends on the missing variable. In contrast to the MAR situation, the missing variable in the NMAR case cannot be predicted only from the available variables in the database. If a sensor cannot acquire information outside a certain range, its data are missing due to NMAR factors. Then the data are said to be censored. If missing data are NMAR, valuable information is lost from the data; and there is a no general method of handling missing data properly.

### **Solution methods**
### **Complete case analysis**, Learn Graph by use of only those data where complete feature information is available. 

Listwise deletion - samples that have missing features are dropped from the analysis of Graph learning model.
Assuming samples are MCAR. This has a **limitation**, that is, if much of data is having missing features than remaining sample is not large enough for learning.



### **Availabe case analysis or pairwise deletion**, Alternative to complete-case analysis, fill in missing observations with plausible values.

Uses all available data to estimate parameters of the model by means of Univariate or Multivariate descriptive statistics of available data.

In **pairwise deletion**, all samples would be used to estimate the mean where complete feature data is available but only the complete cases would contribute to an estimate of mean and variance where missing features are present. 

**Limitation**, When variables are highly correlated, available case analysis provides estimates that are inferior to complete case results, Haitovsky(1968).

Another difficulty is that available case analysis can produce estimated covariance matices that are implaussible, such as etimating correlations outside range of -1.0 to 1.0. Available case analysis will provide consistent estimates only when variables are weakly correlated. Not useful as a general method.

### **Imputation method based on statistical analysis**

1.   **Mean/Mode imputation** Mean value imputation from complete data in case of continous data and Mode in case of Categorical data
2.   **Regression imputation** An advantage of this approach over mean substitution is that it preserves the variance and covariance of variables with missing data. When regression is used to impute missing values on independent variables, this will contribute to multi-colinearity, because the imputed values for missing data will be perfectly correlated with the rest of the variables in the model. The disadvantage of this approach is that all the imputed values follow a single regression curve and cannot represent any inherent variation in the data.
3.   **Hot and cold deck imputation** Hot deck imputation replaces the missing data with the values from a similar complete data vector, i.e., the data vector that is closest in terms of the components that are present in both vectors for each case with a missing value. Hot deck
imputation has the shortcoming that the estimate of the missing data components are based on a single complete vector in the data set, ignoring any global properties of the data set. Cold deck imputation is same as hot deck except that the data source must be different from the current set.

While single value imputation methods as discussed above allows the inclusion of all cases in a standard analysis procedure, replacing missing values with a single value changes the distribution of that variable by decreasing the variance that is likely present, Little (1992). Under no circumstances does mean/mode imputation produce unbiased results.

4.   **Multiple imputation**
Instead of filling in a single value for each missing one, a multiple imputation procedure replaces each missing value with a set of plausible ones that represent the uncertainty about the right value to impute. The missing values are imputed M times to produce M complete data sets using an appropriate model that incorporates random variation. This method, is used for general purpose handling of missing data in multivariate analysis.

### **Imputation methods based on machine learning**
1.   **Imputation with KNN**
The K-nearest neighbor method is a common hot deck method, in which K nearest neighbors are selected from the complete cases, so that they minimize a similarity measure. 
The main drawback of KNN is that the method looks for the most similar cases, through all the data set (in the complete data portion), which implies a high computational cost.

2.   **Multi-layer perceptron imputation**
A basic multi-layer perceptron (MLP) imputation approach consists of training an MLP using only the complete cases as regression model: given d input feature, each incomplete attribute is learned (as output) by means of the remaining complete attributes given as inputs.
In this approach, there is one MLP model per missing variables combination. Depending on the nature of the attributes to be imputed (continous or deiscrete), different error functions (sum of squares error or cross entropy error) are minimized during the training process.
The main disadvantage of MLP is that it requires several MLP models if missing items apprears in several attributes.

3.   **Auto-associative neural network imputation**
An auto-associative neural network (AANN) is a set of neurons that are completely connected in the sense that each neuron receives input form, and sends output to, all the other neurons. First, the network learns from complete cases, in order to replicate all of the imputs as ouputs. Second, when unknown values are detected, the weights are not updated, instead, the missing values are replaced by the network outputs.
4. **Multi-task learning approaches**


###   **Model-Based Methods for Multivariate Normal Missing Data**
Model based methods are based on the added assumption about the distribution of the data and the nature of the missing data mechanism.
Two commonly used methods are,
1.   **Maximum likelihood methods using the EM algorithms**
Maximum likelihood methods for missing multivariate normal data focus on the estimation of the parameters of the observed data, namely the mean vector and variance-covariance matrix. Because of the assumption that the data is multivariate normal, we can utilize the well-known properties of conditional normal distributions to estimate the expected values of the sums and crossproducts of the variables. Using maximum likelihood with the EM algorithm does not result in values for individual missing variables. In the first step the means and the variance-covariance matrix of the variables of interest are estimated, and then these parameter estimates are used to obtain model parameters such as the coefficients of a linear regression model to estimate missing values.

2.   **Multiple imputation with Multivariate Normal Data**
Multiple imputation avoids two of the dif®culties associated with maximum likelihood methods using the EM algorithm. With multiple imputation, a researcher will use standard methods of analysis once imputations are computed, and can easily obtain standard errors of estimates. Though specialized computing is required in multiple imputation, the method provides much more ¯exibility than in the method described in the previous section. Rubin (1987, 1996) and Schafer (1997) provide comprehensive discussions of the statistical theory of multiple imputation, as well as several examples of its use.



### **Missing Data Imputation with Adversarially-trained Graph Convolutional Networks**

Indro Spinelli (2020) formulated the missing data imputation task in terms of a graph denoising autoencoder, where each edge of the graph encodes the siilarity between two patterns. A GNN encoder learns to build intermediate representations for each eample by interleaving classical projection layers and locally combining information between neighbors, while another decoding GNN learns to reconstruct the full imputed dataset fro this intermediate embedding.

###   **Graph convolutional networks for graphs containing missing features**
All of the methods discussed above has two step strategy in which feature filling and graph learning processes are separated. Hibiki Taguchi (2021) suggested GCN based method which adapts GCN to graphs containing missing features. In contrast to traditional strategy, this method integrates the processing of missing features and graph learning within the same neural network architecture. The missing data is represented by Gaussian Mixture Model (GMM) and expected activation of neurons is calculated in the first hidden layer of GCN, while keeping the other layers of the network unchanged. This enables architechture to learn the GMM parameters and network weight parameters in an endto-end manner.

**Different Approaches to Categorical Encoding**

There are multiple ways of handling Categorical variables and two most widely used techniques are:

###1.   Label Encoding
Label encoding is one of the popular processes of converting labels into numeric values in order to make it understandable for machines. For instance, if we have a column of level in a dataset which includes beginners, intermediate and advanced. After applying the label encoder, it will be converted into 0,1 and 2 respectively. 

###2.   One-Hot Encoding
One-Hot Encoding is one of the most widely used encoding methods in ML models. This technique is used to quantify categorical data where it compares each level of the categorical variable to a fixed reference level. It converts a single variable with n observations and x distinct values, to x binary variables with n observations where each observation denotes 1 as present and 0 as absent of the dichotomous binary variable.

Attribute rbc, pc, pcc, ba, htn, dm, cad, appet, pe, ane and the label are categorical variables and all of them can be encoded as one-hot vector. 

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder

def one_hot(df, cols):
    for each in cols:
        dummies = pd.get_dummies(df[each], prefix=each, drop_first=True)
        df = pd.concat([df, dummies], axis=1).drop([each],axis=1)
    return df
    
def label_encoder(df, cols):
    encoders = dict()
    for col_name in cols:
        series = df[col_name]
        label_encoder = LabelEncoder()
        df[col_name] = pd.Series(
            label_encoder.fit_transform(series[series.notnull()]),
            index=series[series.notnull()].index
        )
        encoders[col_name] = label_encoder
    return df
    

cols = ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane']

X = df.iloc[:, :-1]
y = df.iloc[:, -1]
X = label_encoder(X, cols)
```

##Q2: Semi-Supervised Setting. Semisupervised graph learning with label information.

In general, a graph-based semi-supervised learning (SSL) method consists of two key steps. First, a graph is built from all data samples (including both labeled and unlabeled samples) to model the relationships among the points. Then, label information of the labeled samples is propagated to the unlabeled samples over the graph. Though different graph-based SSL methods formulate the label propagation process via different objective functions, all of them share one common assumption (i.e., the cluster assumption), that is, points on the same structure (such as a cluster, a subspace, or a manifold) are likely to have the same label. Since one normally does not have an explicit model for the underlying structures, a graph constructed from the data samples often serves as an approximation to it. Therefore, onstructing a good graph that best captures the essential data structure is critical for all graph-based SSL methods.

Liansheng (2017) proposed to explicitly incorporate label information into the self-representation methods. Specifically by integrating the label information of a subset of the samples into the state-of-the-art self-representation methods, such as the LRR-graph, **by restricting the representation coefficients between labeled points from different classes to be zero**. Intuitively, this information helps prevent the structure-sparsity of the coefficients from being destroyed in challenging real world scenarios, i.e., small signal-to-noise ratio, dependent subspaces and/or nonlinear manifolds.
